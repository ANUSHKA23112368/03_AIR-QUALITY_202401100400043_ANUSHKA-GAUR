```python
a=tf.constant(5)
print(a)

v=tf.constant([]) # Corrected: An empty list can be converted to an empty tensor
print(v)
m=tf.constant([[1,2],[3,4]]) # Corrected: For a 2D tensor, pass a list of lists
print(m)
t3=tf.constant([[[1,2],[3,4]],[[5,6],[7,8]]])
print(t3)
```

```python
a=tf.constant([10,20,30])
b=tf.constant([40,50,60])
print(a+b)
```

```python
s1 = tf.constant(10)
s2 = tf.constant(3)

print("Scalar 1:", s1)
print("Scalar 2:", s2)

# Addition
print("Addition:", s1 + s2)

# Subtraction
print("Subtraction:", s1 - s2)

# Multiplication
print("Multiplication:", s1 * s2)

# Division
print("Division:", s1 / s2)

# Modulo
print("Modulo:", s1 % s2)
```

```python
m1 = tf.constant([[1, 2], [3, 4]])
m2 = tf.constant([[5, 6], [7, 8]])

print("Matrix 1:\n", m1)
print("Matrix 2:\n", m2)

# Matrix Addition
print("\nMatrix Addition:\n", m1 + m2)

# Matrix Subtraction
print("\nMatrix Subtraction:\n", m1 - m2)

# Matrix Multiplication (element-wise multiplication)
print("\nElement-wise Matrix Multiplication:\n", m1 * m2)

# Dot product / Matrix multiplication (tf.matmul)
print("\nDot Product / Matrix Multiplication (tf.matmul):\n", tf.matmul(m1, m2))
```

```python
v1 = tf.constant([1, 2, 3])
v2 = tf.constant([4, 5, 6])

print("Vector 1:", v1)
print("Vector 2:", v2)

# Addition
print("Vector Addition:", v1 + v2)

# Subtraction
print("Vector Subtraction:", v1 - v2)

# Element-wise Multiplication
print("Element-wise Multiplication:", v1 * v2)
```

4. For `Value`, paste the token you copied from GitHub.
5. Ensure `Notebook access` is toggled on.
6. Once saved, you can access this token in your notebook using `user_secrets.get('GITHUB_TOKEN')`.
