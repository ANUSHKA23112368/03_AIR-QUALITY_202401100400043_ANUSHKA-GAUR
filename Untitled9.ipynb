```python
df['Temperature (° F)'] = df['Temperature (° F)'].fillna(df['Temperature (° F)'].mean())
df['Wind speed (MPH)'] = df['Wind speed (MPH)'].fillna(df['Wind speed (MPH)'].mean())
df['Humidity (%)'] = df['Humidity (%)'].fillna(df['Humidity (%)'].mean())
df['Payload Type'] = df['Payload Type'].fillna(df['Payload Type'].mode()[0])
df['Launch Site'] = df['Launch Site'].fillna(df['Launch Site'].mode()[0])
```

```python
df['Mission Status'] = df['Mission Status'].str.lower()
success_data=df[df['Mission Status']=='success']
success_count=success_data['Company'].value_counts()
success_count
```

```python
import matplotlib.pyplot as plt
plt.figure()
success_count.plot(kind='bar')
plt.title('Number of successfull missioms by company')
plt.xlabel('Company')
plt.ylabel('Number of missions')
plt.show()
```

```python
df['Success_Flag'] = df['Mission Status'].apply(lambda x: 1 if x == 'success' else 0)

company_stats = df.groupby('Company').agg(
    Total_Missions=('Mission Status', 'count'),
    Successful_Missions=('Success_Flag', 'sum')
)

company_stats['Success_Rate (%)'] = (company_stats['Successful_Missions'] / company_stats['Total_Missions'] * 100).round(2)

company_stats
```
