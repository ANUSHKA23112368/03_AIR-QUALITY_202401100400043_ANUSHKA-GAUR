```python
import pandas as pd
df = pd.read_csv("/content/air_quality.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Set the style for seaborn
sns.set(style="whitegrid")

# 1. Scatter plot: PM2.5 vs NO2 (color by temperature)
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['pm25'], df['no2'], c=df['temperature'], cmap='coolwarm', alpha=0.7)
plt.colorbar(scatter, label='Temperature (Â°C)')
plt.xlabel('PM2.5')
plt.ylabel('NO2')
plt.title('PM2.5 vs NO2 colored by Temperature')
plt.grid(True)
plt.show()
```

```python
# Function to classify air quality based on PM2.5, NO2, and temperature
def classify_air_quality(pm25, no2, temperature):
    if pm25 <= 12 and no2 <= 53:  # If both PM2.5 and NO2 are in the "Good" range
        return "Good", "Air quality is satisfactory, and air pollution poses little or no risk."
    elif pm25 <= 35.4 and no2 <= 100:  # If PM2.5 and NO2 fall in the "Moderate" range
        return "Moderate", "Air quality is acceptable; however, there may be a risk for some people who are unusually sensitive to air pollution."
    elif pm25 <= 55.4 or no2 <= 360:  # Conditions for "Unhealthy for Sensitive Groups"
        return "Unhealthy for Sensitive Groups", "Members of sensitive groups may experience health effects. The general public is less likely to be affected."
    elif pm25 <= 150.4 or no2 <= 649:  # Conditions for "Unhealthy"
        return "Unhealthy", "Everyone may begin to experience health effects; members of sensitive groups may experience more serious effects."
    else:  # If PM2.5 or NO2 levels are very high
        return "Very Unhealthy", "Health warnings of emergency conditions. The entire population is more likely to be affected."
```

```python
# --- Input section ---
print("ð AIR QUALITY INPUT")  # Print header for input section

pm25 = float(input("Enter PM2.5 level (Âµg/mÂ³): "))  # Take PM2.5 value from the user
no2 = float(input("Enter NO2 level (ppb): "))       # Take NO2 value from the user
temperature = float(input("Enter temperature (Â°C): "))  # Take temperature from the user

```

```python
# --- Classification ---
category, explanation = classify_air_quality(pm25, no2, temperature)  # Call function to classify the air quality
```
